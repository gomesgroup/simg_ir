```python
num_graphs = 16

graphs = []
for i in range(num_graphs):
    num_nodes = torch.randint(low=50, high=100, size=(1,)).item()
    edge_prob = torch.rand(size=(1,)).item()
    graphs.append(Data(x=torch.randn(num_nodes, 39), edge_index=erdos_renyi_graph(num_nodes=num_nodes, edge_prob=edge_prob)))
```

```python
loader = DataLoader(graphs, batch_size=num_graphs)
batch = next(iter(loader))
```
