```python
import wandb
import pandas as pd
api = wandb.Api()

# Get all runs from the sweep
sweep_path = "gomes-group/simg-ir/sweeps/f1hx20gh"
sweep = api.sweep(sweep_path)
runs = sweep.runs

# Create lists to store the hyperparameters and metrics
data = []

# Extract data from each run
for run in runs:
    config = run.config
    data.append({
        'hidden_channels': config['hidden_channels'],
        'num_layers': config['num_layers'],
        'num_ffn_layers': config['num_ffn_layers'],
        'val_loss': run.summary.get('best_val_loss')
    })

# Create dataframe
df = pd.DataFrame(data)
```
